```python
%matplotlib inline
import sys
import random
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
import sympy


print (sys.version)
```

```python
total_num = 100000

x = np.random.exponential(1,total_num)
y = np.random.exponential(x,total_num)

u = ((y - 1) / 3) + 1
t = np.ones(total_num)
```

```python
plt.scatter(x,y,c = 'r') # Show the generated values in red
plt.scatter(t,y,c = 'g') # Show the constant estimator in green
plt.scatter(u,y) # Show the linear estimator in blue


plt.show
```

```python
print("The average value for the constant MSE estimator is:", np.sum(t)/total_num)
print("The average value for the linear MSE estimator is:", np.sum(u)/total_num)
print("The average value for X is:", np.sum(x)/total_num)
print("The average value for Y is:", np.sum(y)/total_num)
```
