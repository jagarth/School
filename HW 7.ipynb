```python
%matplotlib inline
import sys
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
import sympy


#print (sys.version)
```

```python
N = 100000
#Generate the array of uniform random variables
u = np.random.rand(N)
v = np.random.rand(N) 

# Find the marginal distribution of X from the inverse function
x = np.sqrt(1-np.sqrt(1-u))
v = np.sqrt(1-np.sqrt(1-v))

# Find the values of Y dependent upon X from the inverse of the 
# conditional distribution function
y = v / (4*x)

# Find the values of Y from the joint distribution function.
#Y = 8*x*y
#Y = np.sqrt(v/(2*x**2)) 
Y = 4*x*y

#For plotting x and y on the same histogram
z = x,y


#Generate bins for placing values.
bins = np.linspace(0.0, 1.0, num=50)
```

```python
# Plot a histogram of the values of X
#plt.hist(z,bins)
plt.hist(x,bins)
plt.show()
```

The histogram roughly matches the marginal density function fx(x) = -4x(x^2-1). That density function is show below.

```python
# Plot the actual marginal distribution function using sympy
s,t = symbols('s t')
t = -4*s*(s*s-1)
plot(t, (s, 0, 1))
```

```python
#Plot the generated values from the conditional density function. 

plt.hist2d(x,y,bins)
plt.show()
```

```python
#Plot the generated joint density function. 

plt.hist2d(x,Y,bins)
plt.show()
```

This value should roughly match the joint density function fXY(x,y).
