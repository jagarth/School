```python
#!/usr/bin/env python -W ignore::DeprecationWarning
%matplotlib inline
import sys
import random
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
import sympy


print (sys.version)
```

```python
total_num = 10000

u  = np.random.rand(int(total_num/2))
v = np.random.rand(total_num)
y = v
x_plus = 1 + np.sqrt(1 - y[:int(total_num/2)]) * u
x_minus = 1 -  np.sqrt(1 - y[int(total_num/2) :]) * u

x = np.concatenate((x_plus, x_minus), axis=0)



MMSE_l = 6*x-5
MMSE_b = -1*np.log(((x**2) - 2*x + 2)/2)
```

```python
plt.scatter(x,y)
plt.scatter(x,MMSE_l,c = 'r')
plt.scatter(x,MMSE_b, c = 'g')
plt.ylim([0,2])

plt.show
```

```python
print("The average value for the linear MSE estimator of Y is:", np.nansum(MMSE_l)/total_num)
print("The average value for the best MSE estimator of Y is:", np.nansum(MMSE_b)/total_num)
print("The average value for X is:", np.nansum(x)/total_num)
print("The average value for Y is:", np.nansum(y)/total_num)
```

The best estimator is clearly much more effective than the linear estimator. The best estimator takes into account the curvature and is a better approximation for the average value. 

This simulation also shows that the best MSE estimator for Y isn't necessarily the expected value of Y.
