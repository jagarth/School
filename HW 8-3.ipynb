```python
%matplotlib inline
import sys
import random
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
import sympy


print (sys.version)
```

```python

Nx = np.logspace(0, 6, num=30, dtype = int, endpoint = True)

Rx = np.zeros(np.size(Nx))
Ry = np.zeros(np.size(Nx))
Rxy = np.zeros(np.size(Nx))
#Generate the array of uniform random variables

for j in range(np.size(Nx)):
    u = np.random.rand(Nx[j])
    v = np.random.rand(Nx[j]) 

    
    # Make sure Y > X
    for i in range(Nx[j]):
        while u[i] > v[i]:
            u[i] = random.uniform(0,1)
            v[i] = random.uniform(0,1)   
    
    #Get the values for X and Y from the Marginals
    Y = v / 5
    X = np.sqrt(2*u / 15)
    Rx[j] = np.sum(np.multiply(X,u))/Nx[j] 
    Ry[j] = np.sum(np.multiply(Y,v))/Nx[j] 
    
    Rxy[j] = np.sum(np.multiply(X+Y,u+v))/Nx[j] 
```

Part A. Draw variables from the distribution of X and find the expexted value of X as a function of N draws.

```python

plt.axhline(y=0.084, color='r')
plt.semilogx(Nx,Rx)
```

Part A. Draw variables from the distribution of Y and find the expexted value of Y as a function of N draws.

```python

plt.axhline(y=0.1, color='r')
plt.semilogx(Nx,Ry)
```

Part A. Draw variables from the joint distribution of X+Y and find E[X+Y] and compare it to E[X] + E[Y]

```python
plt.semilogx(Nx,Rx+Ry)
plt.semilogx(Nx,Rxy, color='g')
```

The results of E[X+Y] and E[X] + E[Y] show some sort of linear relationship, but not exactly the same. This discrepency is expected because the values of X and Y are dependent with one another this effects their expected values. Therefore, if X and Y are dependent random variables E[X+Y]  E[X] + E[Y].
